#Inport Libraries

```python
import os.path
from transformers import ViTFeatureExtractor, ViTForImageClassification
from hugsvision.inference.VisionClassifierInference import VisionClassifierInference
import matplotlib.image as mpimg
from matplotlib.pyplot import imshow
```

#Trained Weights Path & Model Loading

```python
path = "/content/huge_weights/model"
feature_extractor="/content/huge_weights/feature_extractor"
classifier = VisionClassifierInference(
    feature_extractor = ViTFeatureExtractor.from_pretrained(feature_extractor),
    model = ViTForImageClassification.from_pretrained(path),
)
```

#Test Image (Select any Image from huge_weights/Test_Smaples directory) Or Test with your own Image after to upload in Local Cloab.

```python
img  = "/content/huge_weights/Test_Samples/diffusion_1.jpg"
```

#Prediction

```python
label = classifier.predict(img_path=img)
print('\033[91m' + 'Predicted Class: '+ '\033[92m'+label)
%matplotlib inline
testim = mpimg.imread(img)
imshow(testim)
```
